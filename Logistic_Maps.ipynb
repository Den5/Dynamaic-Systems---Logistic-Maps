I have made a simple Python script with the functionaility as the NetLogo model provided on the course, "LogisticModel".
Some models with fix point attractors, and some period attractors. Then finaly a model with "Sensitive Dependence on Initial Conditions", Chaos.

```python
import numpy as np
import sys
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
```

```python
# Some Logic Maps xt+1 = R X (1 - xt)
```

```python
R = float(input ('what is the value of R0?' ))
X = float(input ('what is the vale of x?'))
t = int(input ('How my interations(t) ?' ))
```

Creating a function to run the logic map formula

```python
def Logic_Map(X , R, t):
    
    out = np.array([])
    while t > 0:
        X = R *  X *(1 - X)
        out = np.append(out, [X])
        t = t - 1
         
    return out
```

```python
graph = (Logic_Map(X, R, t))
print (graph)
```

```python
plt.plot(graph)
plt.title('Logic Map')
plt.ylabel('X(t)')
plt.xlabel('t')
plt.show
```

Lets create a Logic map with X = 0.5 , R = 2 and for 20 Interations (t)

```python
graph = (Logic_Map(0.5, 2, 20))
plt.title('Logic Map')
plt.ylabel('X(t)')
plt.xlabel('t')
plt.plot(graph)
plt.ylim(ymin=0)  
plt.xlim(xmin=0)
plt.show
```

Lets create a Logic map with X = 0.2 , R = 2 and for 20 Interations (t)


```python
graph = (Logic_Map(0.2, 2, 20))
plt.title('R = 0.2')
plt.ylabel('X(t)')
plt.xlabel('t')
plt.plot(graph)
plt.show
```

This is an attractor, i.e the intial condition is attracted to the value 5.0

```python
graph = (Logic_Map(0.99, 2, 20))
plt.title('R = 2')
plt.ylabel('X(t)')
plt.xlabel('t')
plt.plot(graph)
plt.show
```
